```python
import os
import pandas as pd
os.chdir("C:/Users/Akhil/Desktop")
train = pd.read_csv("Train_Dataset.csv")
test = pd.read_csv("Test_Dataset.csv")
```

```python
train.head()
```

```python
train['Exp'] = train["experience"].str.extract("(\d*\d.?\d+)", expand=True)
```

```python
train['Exp'] = train['Exp'].str.split('-')
```

```python
train.head()
```

```python
train[['start','end']] = pd.DataFrame(train.Exp.values.tolist())
```

```python
train['Avg Exp'] = train[['start','end']].astype(float).mean(axis=1)
```

```python
train['Salary'] = train['salary'].str.split('to')
```

```python
train[['sal_start','sal_end']] = pd.DataFrame(train.Salary.values.tolist())
```

```python
train['Avg_Salary'] = train[['sal_start','sal_end']].astype(float).mean(axis=1)
```

```python
train.head()
```

```python
X = train[['job_desig', 'key_skills', 'City coded', 'Avg Exp', 'Avg_Salary']]
```

```python
X.drop(['Avg Exp'], axis=1)
```

```python
X.dropna(subset=['key_skills'], inplace=True)
```

```python
X.key_skills.isnull().sum()
```

```python
X.head()
```

```python
# Import Pipeline
from sklearn.pipeline import Pipeline

# Import other necessary modules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[['Avg Exp']],
                                                    X[['Avg_Salary']], 
                                                    random_state=22)
```

```python
pl = Pipeline([
        ('clf', OneVsRestClassifier(LogisticRegression()))
    ])
```

```python
pl.fit(X_train, y_train)
```

```python
accuracy = pl.score(X_test, y_test)
```

```python
print("\nAccuracy: ", accuracy)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import FunctionTransformer
```

```python
get_text_data = FunctionTransformer(lambda x: x[['job_desig', 'key_skills', 'City coded']], validate=False)
```

```python
get_numeric_data = FunctionTransformer(lambda x: x[['Avg Exp']], validate=False)
```

```python
just_text_data = get_text_data.fit_transform(X)
```

```python
just_numeric_data = get_numeric_data.fit_transform(X)
```

```python
from sklearn.pipeline import FeatureUnion
```

```python
X_train, X_test, y_train, y_test = train_test_split(X[['job_desig', 'key_skills', 'City coded', 'Avg Exp']],
                                                    X[['Avg_Salary']], 
                                                    random_state=22)
```

```python
process_and_join_features = FeatureUnion(
            transformer_list = [
                ('numeric_features', Pipeline([
                    ('selector', get_numeric_data)
                ])),
                ('text_features', Pipeline([
                    ('selector', get_text_data),
                    ('vectorizer', CountVectorizer())
                ]))
             ]
        )

# Instantiate nested pipeline: pl
pl = Pipeline([
        ('union', process_and_join_features),
        ('clf', OneVsRestClassifier(LogisticRegression()))
    ])
```

```python
pl.fit(X_train, y_train)
```
